# If you want to express yourself with graph plotting
Jupyter Notebook & SageMath Kernel

```python
@interact
def _(n=(30,75,5)):
    g = graphs.CubeGraph(4); edge_colors = {}; c = [colormaps.hsv_r(n*i)[:3] for i in [0..3]]
    for i in [0..3]:
        edge_colors[c[i]] = []
    for u,v,l in g.edges():
        [edge_colors[c[i]].append((u,v,l)) for i in [0..3] if u[i] != v[i]] 
    g.graphplot(edge_colors=edge_colors,edge_thickness=2,
                vertex_labels=False,vertex_size=0).show(figsize=(5,5))
```

# When you want to draw with graphs
Jupyter Notebook & SageMath Kernel

```python
@interact
def _(n=[3,4,5,6]):
    g = Graph({},loops=True,multiedges=True,sparse=True)
    def add(g,x,y):
        g.add_edges([(x,y,'a'),(x,y,'a'),(x,y,'b'),(x,y,'b'),(x,y,'c'),(x,y,'c'),
                     (x,y,'d'),(x,y,'d'),(x,y,'e'),(x,y,'e'),(x,y,'f'),(x,y,'f')])
    [add(g,i,i+1) for i in [1..n-1]]; add(g,n,1)
    g.graphplot(vertex_size=0,vertex_labels=False,color_by_label=True).show(figsize=(5,5))
```

# If you think it's hard to draw reflections. Think again
Jupyter Notebook & SageMath Kernel

```python
@interact
def _(N=(1,5,1)):
    var('x,y,z')
    p1=sum([implicit_plot3d(x^2+y^2-2*z*i^2,(x,-i,i),(y,-i,i),(z,0,1),color=hue(sin(0.24*i)),
                            opacity=0.7,plot_points=5,frame=False) for i in [1..N]])
    p2=sum([implicit_plot3d(x^2+y^2+2*z*i^2,(x,-i,i),(y,-i,i),(z,-1,0),color=hue(sin(0.24*i)),
                            opacity=0.3,plot_points=5,frame=False) for i in [1..N]])
    p3=plot3d(lambda x,y:0,(-2*N,2*N),(-2*N,2*N),color="steelblue",opacity=0.5)
    (p1+p2+p3).show()
```
