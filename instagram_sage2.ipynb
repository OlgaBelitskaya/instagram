# When you want to draw with math formulas
Jupyter Notebook & SageMath Kernel

```python
var('u'); R,r,n=12,6,24; x=r*cos(u); y=r*sin(u); 
p1=sum([parametric_plot((x-R*cos(2*i*pi/n),y-R*sin(4*i*pi/n)),(u,0,2*pi),color='#3636ff') for i in [0..n]])
p2=sum([parametric_plot((y-R*sin(4*i*pi/n),x-R*cos(2*i*pi/n)),(u,0,2*pi),fill=True,fillalpha=0.1,
                        fillcolor=hue(i/n),color='#3636ff') for i in [0..n]])
p3=sum([parametric_plot((1.5*x-R*cos(2*i*pi/n),1.5*y-R*sin(2*i*pi/n)),(u,0,2*pi),color='darkblue') for i in [0..n]])
(p1+p2+p3).show(figsize=[7,7],axes=False)
```
