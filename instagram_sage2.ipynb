# When you want to imagine what is the cross product
Jupyter Notebook & SageMath Kernel

```python
x,y,z = 2,3,4; av = vector(SR,[x,y,z]); bv = vector(SR,[x,y,-z]); cv = av.cross_product(bv)
a=arrow3d((0,0,0),av,color='#ff3636',width=(x+y+z)/1.5)+text3d('a {x,y,z}',av*1.1)
b=arrow3d((0,0,0),bv,color='#3636ff',width=(x+y+z)/1.5)+text3d('b {x,y,-z}',bv*1.1)   
c=arrow3d((0,0,0),cv,color='#ff36ff',width=(x+y+z)/1.5)+text3d('axb',cv*1.1)
(a+b+c).show()
```

# If you want to express yourself with graph plotting
Jupyter Notebook & SageMath Kernel

```python
@interact
def _(n=(30,75,5)):
    g = graphs.CubeGraph(4); edge_colors = {}; c = [colormaps.hsv_r(n*i)[:3] for i in [0..3]]
    for i in [0..3]:
        edge_colors[c[i]] = []
    for u,v,l in g.edges():
        [edge_colors[c[i]].append((u,v,l)) for i in [0..3] if u[i] != v[i]] 
    g.graphplot(edge_colors=edge_colors,edge_thickness=2,
                vertex_labels=False,vertex_size=0).show(figsize=(5,5))
```
